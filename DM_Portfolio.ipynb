### Normalizing the data

In order to get more insights from the scater plot chart, we have to normalize/scale the data. We'll use 0,1 scaler to obtain this task or the lognorm function from numpy, and based on the results will decide which is better after handling the missing values.

```python
# Scaler function
def scaler(Y, y, min=0, max=1):
    y_scaled = min + (Y - y.min() * (max-min)) / (y.max() - y.min())
    return y_scaled
```

```python
#Lognorm function
def lognorm(y):
     y_log = np.log1p(y)
     return y_log
```

```python
"""
Feature Selection 2 - Mutual Information Classification

    * Mutual information (MI) between two random variables is a non-negative value, 
      which measures the dependency between the variables. 

    * It is equal to zero if and only if two random variables are independent, 
      and higher values mean higher dependency.

    * The function relies on nonparametric methods based on entropy estimation from 
      k-nearest neighbors.
"""

from sklearn.feature_selection import mutual_info_classif
dependentFeature = [['Response'],[df['Response']]]

MutualSet = df.loc[:, df.columns != 'Response']

def findDependencyFeatures(name,labelset):
    mutual_info = mutual_info_classif(MutualSet,labelset)
    mutual_info = pd.Series(mutual_info)
    mutual_info.index = MutualSet.columns
    ax = mutual_info.sort_values(ascending=False).plot.bar(figsize=(15,5))
    ax.set_ylabel(name)
    plt.show()

for i in range(len(dependentFeature[0])):
    findDependencyFeatures(dependentFeature[0][i],dependentFeature[1][i])
```

```python
"""
Select Top Perecentile
"""
from sklearn.feature_selection import SelectPercentile
top_features = SelectPercentile(mutual_info_classif,percentile=75)
top_features.fit(MutualSet,df['Response'])

print(f"Top 75% Features: \n\n{MutualSet.columns[top_features.get_support()]} \n\n and these are {len(MutualSet.columns[top_features.get_support()])} out of {len(df.columns)} ")
```

```python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision", linewidth=2)
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall", linewidth=2)
    plt.legend(loc="center right", fontsize=16)  
    plt.xlabel("Threshold", fontsize=16)        
    plt.grid(True)                              
    plt.axis([-5, 5, 0, 1])             



recall_90_precision = recalls[np.argmax(precisions >= 0.90)]
threshold_90_precision = thresholds[np.argmax(precisions >= 0.90)]


plt.figure(figsize=(8, 4))                                                                  
plot_precision_recall_vs_threshold(precisions, recalls, thresholds)
plt.plot([threshold_90_precision, threshold_90_precision], [0., 0.9], "r:")                 
plt.plot([-5, threshold_90_precision], [0.9, 0.9], "r:")                                
plt.plot([-5, threshold_90_precision], [recall_90_precision, recall_90_precision], "r:")
plt.plot([threshold_90_precision], [0.9], "ro")                                             
plt.plot([threshold_90_precision], [recall_90_precision], "ro")                                                                       # Not shown
plt.show()
```

```python
def plot_precision_vs_recall(precisions, recalls):
    plt.plot(recalls, precisions, "b-", linewidth=2)
    plt.xlabel("Recall", fontsize=16)
    plt.ylabel("Precision", fontsize=16)
    plt.axis([0, 1, 0, 1])
    plt.grid(True)

plt.figure(figsize=(8, 6))
plot_precision_vs_recall(precisions, recalls)
plt.plot([recall_90_precision, recall_90_precision], [0., 0.9], "r:")
plt.plot([0.0, recall_90_precision], [0.9, 0.9], "r:")
plt.plot([recall_90_precision], [0.9], "ro")
plt.show()
```
